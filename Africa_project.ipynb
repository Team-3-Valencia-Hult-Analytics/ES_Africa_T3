```python
# Africa

Africa = ['AGO','BWA','BDI','COG','DJI','ERI', 'SWZ','ETH','KEN','LSO','MDG', 
          'MWI','MOZ','NAM','RWA','SOM','ZAF','SSD','TZA','UGA','ZMB',
          'ZWE']

# Subset of Countr - dies
AfricaDf = Africa_dataset[Africa_dataset['Country Code'].isin(Africa)]

AfricaDf.rename(columns={"Country Name":"Country_Name","Employment in agriculture (% of total employment) (modeled ILO estimate)"
                         : "Emp_agriculture","Employment in industry (% of total employment) (modeled ILO estimate)"
                         : "Emp_industry","Employment in services (% of total employment) (modeled ILO estimate)"
                         : "Emp_services","GDP (current US$)":"GDP_USD","GDP growth (annual %)":"GDP_growth"}, inplace=True)

AfricaDf.head() #loc[:,["Emp_agriculture"]]

```

```python

# Checking missing values for Africa Data Set
AfricaT = AfricaDf.transpose()

AfricaT.columns = Africa

# mv = missing values

Africa_mv_sorted = (AfricaT.isnull().sum().sort_values(ascending = False) / len(AfricaT)) * 100

Africa_mv_sorted
```

```python
#UNCOMMENT (!PIP INSTALL) BEFORE RUNNING THIS CODE AND THEN COMMENT BACK AGAIN TO SPEED THE PROCESS
#!pip install world_bank_data --upgrade 

import world_bank_data as wb

# It's necessary to get the indicators code, this is different from the column names. 
    # E.g: Income share held by fourth 20% is coded as: SI.DST.04TH.20
    
missingIndicators = Africa_mc[Africa_mc >= 1].index.copy()

# Created empty dataframe with the column format that the API will return
    # We just need to append in a simple loop
indicatorsCodedMissing = pd.DataFrame(columns = ['name', 'unit', 'source', 'sourceNote', 'sourceOrganization', 'topics'])

# Loop to get the missing indicators and store the in the indicatorsCodedMissing data frame
for i in list(missingIndicators):
    indicatorsCodedMissing = indicatorsCodedMissing.append(wb.search_indicators(i))

# view the first the indicator names
print(f'{indicatorsCodedMissing.index[0:3]} \n')

indicatorsCodedMissing.head()
```

```python
# Now we obtain the data for the specific countries that we need

#UNCOMMENT (!PIP INSTALL) BEFORE RUNNING THIS CODE AND THEN COMMENT BACK AGAIN TO SPEED THE PROCESS 
#!pip install wbdata

import wbdata
from datetime import datetime

# missingIndicators = AfricaheadMissingCols[AfricaheadMissingCols > 1] 

# The indicators argument of get_dataframe needs to be a dictionary with 
    # the codes (previous step) as the key and the new column names as the values (missing Indicators)
    # It will return the Data frame with all the years in the world bank 
    
keys = list(indicatorsCodedMissing.index)
values = list(missingIndicators) 

wbQueriedAfrica = wbdata.get_dataframe(indicators = dict(zip(keys,values)), country = Africa, convert_date = False)

wbQueriedAfrica.head()
```

```python
# Subseting  for a particular year

wbQueriedAfrica.sort_index(inplace=True)

# What is the total of missing values that there are per year 
for i in range(2010,2019,1):
    print('{} missing values out of {} for year {} '.format(wbQueriedAfrica.xs(str(i),level='date'). 
                                                            transpose().isnull().sum().sum(), # Calculates overall sum of null values
                                                            wbQueriedAfrica.shape[1] * len(wbQueriedAfrica.index.levels[0]), # Number of columns * number of countries = total possible missing values 
                                                            i) # ith year 
        )

# It seems that 2015 is the most non-missing updated info that we can obtain 

# data for 2015 is queried
wbQueriedAfrica2017 = wbQueriedAfrica.xs('2017',level='date')

# Printing the data frame to know the contents
wbQueriedAfrica2017

# saving it to excel
   
wbQueriedAfrica2017.to_excel('worldBank2017AfricaCountries.xlsx')  # This is commented so I don't overwrite by mistake the file
```
